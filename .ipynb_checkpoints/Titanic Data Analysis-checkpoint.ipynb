```python
def male_female_child(passenger):
    age,sex = passenger
    if age < 16:
        return 'child'
    else:
        return sex

titanic_df['Person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
deck = titanic_df['Cabin'].fillna('U')
deck.head()
```

```python
levels = []
for item in deck:
    levels.append(item[0])

cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']
cabin_df = cabin_df[cabin_df.Cabin != 'T']
sns.factorplot('Cabin',data=cabin_df,kind='count')
```

```python
sns.factorplot('Embarked',data=titanic_df,hue='Pclass',kind='count')
```

```python
titanic_df['Alone'] = titanic_df.Parch + titanic_df.SibSp
titanic_df.head()
```

```python
titanic_df['Alone'].loc[titanic_df['Alone'] > 0] = 'With family'
titanic_df['Alone'].loc[titanic_df['Alone'] == 0] = 'Alone'
titanic_df.head()
```

```python
titanic_df['Survivor'] = titanic_df['Survived'].map({0:'no',1:'yes'})
sns.factorplot('Survivor',data=titanic_df,kind='count')
```

```python
def get_status(name):
    a = name.split(",")[1]
    return a.split(".")[0].strip()

Title_Dictionary = {
                    "Capt":       "Officer",
                    "Col":        "Officer",
                    "Major":      "Officer",
                    "Jonkheer":   "Royalty",
                    "Don":        "Royalty",
                    "Sir" :       "Royalty",
                    "Dr":         "Officer",
                    "Rev":        "Officer",
                    "the Countess":"Royalty",
                    "Dona":       "Royalty",
                    "Mme":        "Mrs",
                    "Mlle":       "Miss",
                    "Ms":         "Miss",
                    "Mr" :        "Mr",
                    "Mrs" :       "Mrs",
                    "Miss" :      "Miss",
                    "Master" :    "Master",
                    "Lady" :      "Royalty"

                    }

titanic_df['Status'] = titanic_df['Name'].apply(get_status).map(Title_Dictionary)
titanic_df.head()
```

```python
titanic_df['Status'].value_counts()
```

```python
sns.factorplot('Status','Survived',data=titanic_df)
```

```python
titanic_df['Deck'] = cabin_df['Cabin']
sns.factorplot('Deck','Survived',data=titanic_df)
```

```python
def ticket_type(t):
    t = t.replace( '.' , '' )
    t = t.replace( '/' , '' )
    t = map( lambda ti : ti.strip() , t )
    t = list(filter( lambda ti : not ti.isdigit() , t ))
    if len( t ) > 0:
        return t[0]
    else: 
        return 'XXX'
    
titanic_df['TicketClass'] = titanic_df['Ticket'].apply(ticket_type)
sns.factorplot('TicketClass','Survived',data=titanic_df)
```
